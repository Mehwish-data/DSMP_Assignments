Python Visualization (Matplotlib/Seaborn):

Use the same dataset to:

Create a histogram to understand data distribution.

scatter plot to showcase relationships between two numerical variables.

heatmap to highlight correlations.


```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
url = "https://raw.githubusercontent.com/dataseekho/sql-beyond-basics/main/dataset/Sales_Large.csv"
df = pd.read_csv(url)

# Show all column names
print(df.columns)
```

```python
# Preview the data
print(df.head())
```

```python
# 1. Histogram: Distribution of Revenue
plt.figure(figsize=(8, 5))
sns.histplot(df['Revenue'], bins=50, kde=True, color='skyblue')
plt.title("Distribution of Revenue")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2. Scatter Plot: Revenue vs Discounts
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Revenue', y='Discounts', color='green')
plt.title("Revenue vs Discounts")
plt.xlabel("Revenue")
plt.ylabel("Discounts")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 3. Heatmap: Correlation between Revenue, Discounts, and Taxes
plt.figure(figsize=(8, 6))
corr = df[['Revenue', 'Discounts', 'Taxes']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
