To introduce **unsupervised learning** concepts like **clustering** and **dimensionality reduction,**
 Analyzing the "**Wholesale Customers Data Set**." This dataset includes annual spending amounts of clients from a wholesale distributor across various product categories.


**Task 1: Data Preprocessing: Handle missing values, detect and treat outliers, and normalize the data as necessary.**

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler

#Load dataset(replace'your_data.csv' with actual data file path)
df=pd.read_csv("/content/Wholesale customers data.csv")
# Handling Missing Values
imputer=SimpleImputer(strategy='mean')  # You can change to 'median' or 'most_frequent'
df_imputed = pd.DataFrame(imputer.fit_transform (df.select_dtypes(include=['number'])),columns=df.select_dtypes(include=['number']).columns)
# Detecting and treating Outliers using Z-score
z_scores = np.abs((df_imputed - df_imputed.mean()) / df_imputed.std())
outlier_threshold = 3  # Typically, values above 3 are considered outliers
df_no_outliers = df_imputed[(z_scores < outlier_threshold).all(axis=1)]

# Normalizes the data(choose one method)
scaler = StandardScaler() # Standardization (mean=0, std=1)
df_standardized = pd.DataFrame(scaler.fit_transform(df_no_outliers), columns=df_no_outliers.columns)

# OR
# scaler = MinMaxScaler()  # Min-Max Scaling (0-1 range)
# df_normalized = pd.DataFrame(scaler.fit_transform(df_no_outliers), columns=df_no_outliers.columns)

# Display processed data
print(df_standardized.head())
```

**Task 2.Exploratory Data Analysis (EDA): Investigate the distribution of variables and relationships between them using visualizations and summary statistics.**

```python
import matplotlib.pyplot as plt
import seaborn as sns
#Exploratory Data Analysis (EDA)
print("Summary Statistics:")
print(df_standardized.describe())
#Visualizing distribution
plt.figure(figsize=(8,6))
df_standardized.hist(bins=30,figsize=(8,6))
plt.show()
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_standardized.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
#Pairplot for relationships
sns.pairplot(df_standardized)
plt.show()
#Display processed data
print(df_standardized.head())


```

**Task 3: Clustering: Implement K-Means clustering to identify distinct customer segments. Determine the optimal number of clusters using methods like the elbow method or silhouette score.**

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
# Clustering: K-Means
# Determine optimal number of clusters using the Elbow Method
wcss = []  # Within-cluster sum of squares
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_standardized)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal K')
plt.show()

# Determine optimal clusters using Silhouette Score
silhouette_scores = {}
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(df_standardized)
    silhouette_scores[k] = silhouette_score(df_standardized, cluster_labels)

best_k = max(silhouette_scores, key=silhouette_scores.get)
print(f"Optimal number of clusters based on silhouette score: {best_k}")

# Apply K-Means with optimal K
kmeans = KMeans(n_clusters=best_k, random_state=42)
df_standardized['Cluster'] = kmeans.fit_predict(df_standardized)

# Visualizing clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_standardized.iloc[:, 0], y=df_standardized.iloc[:, 1], hue=df_standardized['Cluster'], palette='viridis')
plt.title('K-Means Clustering')
plt.show()

# Display processed data with clusters
print(df_standardized.head())

# Cluster Interpretation
cluster_summary = df_standardized.groupby('Cluster').mean()
print("Cluster Characteristics:")
print(cluster_summary)

# Visualizing Cluster Characteristics
plt.figure(figsize=(12, 6))
sns.heatmap(cluster_summary, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Cluster Characteristics Heatmap")
plt.show()

# Count of data points in each cluster
cluster_counts = df_standardized['Cluster'].value_counts()
print("Cluster Size Distribution:")
print(cluster_counts)

plt.figure(figsize=(8, 5))
sns.barplot(x=cluster_counts.index, y=cluster_counts.values, palette='viridis')
plt.xlabel("Cluster")
plt.ylabel("Number of Data Points")
plt.title("Cluster Size Distribution")
plt.show()
```
