### Remove **HTML tags**, **special characters**, and **extra spaces**.

```python
import re
import pandas as pd
from bs4 import BeautifulSoup

# Load dataset
df = pd.read_csv("/content/IMDB Dataset.csv")

def clean_text(text):
    # Remove HTML tags
    text = BeautifulSoup(text, "html.parser").get_text()

    # Remove special characters, numbers, and punctuation (keeping only letters and spaces)
    text = re.sub(r"[^a-zA-Z\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    return text

df["cleaned_review"] = df["review"].apply(clean_text)

# Display first 20 rows
df.head()
```


*   Convert text to **lowercase** and remove **stopwords**.

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Ensure the necessary NLTK resources are downloaded
nltk.download('punkt')
nltk.download('stopwords')

# Load your dataset (replace 'your_data.csv' with your actual file path)
df = pd.read_csv('/content/IMDB Dataset.csv')

# Define a set of English stopwords
stop_words = set(stopwords.words('english'))

# Function to preprocess text: convert to lowercase and remove stopwords
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Tokenize the text
    words = word_tokenize(text)
    # Remove stopwords
    filtered_words = [word for word in words if word.isalnum() and word not in stop_words]
    # Join words back into a single string
    return ' '.join(filtered_words)
df['processed_text'] = df['review'].apply(preprocess_text)

# Display the first 10 rows to verify the changes
print(df[['review', 'processed_text']].head())

```

 *   Apply **stemming** (reduce words to root form).

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

#  necessary NLTK packages are downloaded
nltk.download("punkt", force=True)
nltk.download("stopwords", force=True)

# Initialize Porter Stemmer
ps = PorterStemmer()

# Load IMDb Dataset
df = pd.read_csv("/content/IMDB Dataset.csv", encoding="utf-8")

# Display first few rows
print(df.head())

# Define stopwords set
stop_words = set(stopwords.words("english"))

# Define text preprocessing function with stemming
def preprocess_text(text):
    # Tokenization
    words = word_tokenize(text)
    # Remove stopwords
    words = [word for word in words if word not in stop_words and word.isalnum()]
    # Apply stemming
    words = [ps.stem(word) for word in words]
    return words

# Apply preprocessing to the "review" column
df["processed_review"] = df["review"].apply(preprocess_text)

# Display processed output
print(df[["review", "processed_review"]].head())
```

*   Visualize word **frequency using bar plots**.

```python
import matplotlib.pyplot as plt
from collections import Counter

# Step 1: Read the file and extract words
try:
    with open('/content/IMDB Dataset.csv', 'r') as file:
        text = file.read()
    words = text.split()
except FileNotFoundError:
    print("The file '/content/IMDB Dataset.csv' was not found. Please check the file path.")
    words = []

# Step 2: Count word frequencies
if words:
    word_counts = Counter(words)
    most_common_words = word_counts.most_common(20)
else:
    most_common_words = []
    print("No words to process.")

# Step 3: Prepare data for plotting
if most_common_words:
    common_words, counts = zip(*most_common_words)
else:
    common_words, counts = [], []

# Step 4: Plot the data
if common_words and counts:
    plt.figure(figsize=(10, 8))
    plt.barh(common_words, counts, color='skyblue')
    plt.xlabel('Frequency')
    plt.title('Top 20 Most Common Words')
    plt.gca().invert_yaxis()
    plt.show()
else:
    print("No data available to plot.")
```


* Generate word clouds for **common words in positive and negative reviews.**

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords

# Download stopwords if not already available
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Load the dataset (Replace 'your_file.csv' with the actual filename)
df = pd.read_csv("/content/IMDB Dataset.csv")

# Check the first few rows to verify column names
print(df.head())

# Ensure correct column names (modify 'review' and 'sentiment' if needed)
if 'review' not in df.columns or 'sentiment' not in df.columns:
    raise ValueError("Check the column names in your dataset!")

# Convert reviews to lowercase and remove NaN values
df['review'] = df['review'].astype(str).str.lower()
df = df.dropna(subset=['review', 'sentiment'])

# Separate positive and negative reviews
positive_reviews = " ".join(df[df['sentiment'] == 'positive']['review'])
negative_reviews = " ".join(df[df['sentiment'] == 'negative']['review'])

# Function to generate word clouds
def generate_wordcloud(text, title):
    wordcloud = WordCloud(width=800, height=400, background_color='white',
                          stopwords=stop_words, colormap='coolwarm').generate(text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(title, fontsize=14)
    plt.show()

# Generate word clouds
generate_wordcloud(positive_reviews, "Common Words in Positive Reviews")
generate_wordcloud(negative_reviews, "Common Words in Negative Reviews")
```

# 3. **Feature Engineering:**Convert text into numbers using:
*  **Bag of Words (BoW)**- Count word occurrences.
*  **TF-IDF** - Give importance to unique words.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
# Load dataset
df = pd.read_csv("/content/IMDB Dataset.csv")

# Extract text data
documents = df['review'].astype(str).tolist()

# Bag of Words (BoW)
bow = CountVectorizer(stop_words='english')
X_bow = bow.fit_transform(documents)

# Print results
print("BoW Matrix Shape:", X_bow.shape)  # Rows = documents, Columns = unique words
print("\nSample Feature Names:", bow.get_feature_names_out()[:40])  # Print first 40 words
print("\nSample BoW Representation:\n", X_bow.toarray()[:20])  # Print first 20 rows
print("BoW Shape:", X_bow.shape)
```

*   **TF-IDF** - Give importance to unique words

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Load dataset
df = pd.read_csv("/content/IMDB Dataset.csv")

# Extract text data
documents = df['review'].astype(str).tolist()

# TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
X_tfidf = tfidf.fit_transform(documents)

# Print results
print("TF-IDF Shape:", X_tfidf.shape)  # Rows = documents, Columns = unique words
print("\nSample Feature Names:", tfidf.get_feature_names_out()[:20])  # First 20 words
print("\nSample TF-IDF Representation:\n", X_tfidf.toarray()[:20])  # First 20 rows
```

# 4. **Build a Classification Model:**
*   Split data into training and testing sets (80-20 split)
*   Train a Nave Bayes classifier to predict sentiment.
*   Evaluate the model using accuracy, precision, and recall.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Load dataset
df = pd.read_csv("/content/IMDB Dataset.csv")

# Convert sentiment labels to binary (positive = 1, negative = 0)
df['sentiment'] = df['sentiment'].map({'positive': 1, 'negative': 0})

# Extract text and labels
documents = df['review'].astype(str).tolist()
labels = df['sentiment']

# Convert text data into numerical format using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)

# Split data (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

# Train Nave Bayes Model
model = MultinomialNB()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate Model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

# Print Evaluation Metrics
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
```
